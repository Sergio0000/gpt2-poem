```python
import re
import pandas as pd
```

```python
archivo = '/content/drive/MyDrive/sis330/dataset/poemas_f.txt'

characters = "“=/\'—¡!¿?,;:'-<>>.0123456789()»«\n\""
versos=[]
with open(archivo) as fname:
  print(fname)
  for linea in fname:
    linea=linea.rstrip()
    linea=linea.lstrip()
    if len(linea) > 0 and len(linea) <1000:
      versos.append(linea)
```

```python
for i in range(len(versos)):
    versos[i] = re.sub("[0-9]", "", versos[i])
    versos[i] = re.sub(r'\[[^\]]+\]', '', versos[i])
    versos[i] = re.sub(' +', ' ', versos[i])
    versos[i] = re.sub(' ― | – ', ' ', versos[i])
    versos[i] = re.sub(' ―|«|»|…|… |:|\;|\:|\/|\(|\)|\<|\>', '', versos[i])
    versos[i] = re.sub(r"^\s+", "", versos[i])
    versos[i] = re.sub('― ', ' ', versos[i])
    versos[i] = re.sub('\n{3,}', '\n\n', versos[i])
```

```python
def valores(a,b,c,d):
  len_ter=3
  ult_tres_p1 = a[-len_ter:] 
  ult_tres_p2 = b[-len_ter:]
  ult_tres_p3 = c[-len_ter:] 
  ult_tres_p4 = d[-len_ter:]

  if (ult_tres_p1 == ult_tres_p3 and ult_tres_p2 == ult_tres_p4 ):
      return 'abab'
  elif ult_tres_p1 == ult_tres_p2 and ult_tres_p3 == ult_tres_p4 :
      return 'aabb'
  elif ult_tres_p1 == ult_tres_p4 and ult_tres_p2 == ult_tres_p3 :
      return 'abba'
  elif ult_tres_p1 == ult_tres_p4 and ult_tres_p2 == ult_tres_p3 :
      return 'baab'
  elif ult_tres_p1 == ult_tres_p4 and ult_tres_p2 == ult_tres_p3 :
      return 'aaab'
  
  return 'x'


lista=[]
rima=[]
for i in range(0,len(versos)-3):
  tipo= valores(versos[i],versos[i+1],versos[i+2],versos[i+3])
  if(tipo !='x'):
    print(valores(versos[i],versos[i+1],versos[i+2],versos[i+3]))
    print(versos[i]+'\n'+versos[i+1]+'\n'+versos[i+2]+'\n'+versos[i+3])
    lista.append(versos[i]+'.\n'+versos[i+1]+'.\n'+versos[i+2]+'.\n'+versos[i+3]+'.\n')
    rima.append(tipo)
```

```python
data = pd.DataFrame(data={"poems": lista,"tipo_rima":rima})
```
